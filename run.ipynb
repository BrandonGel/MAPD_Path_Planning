```python
import csv
import numpy as np
# Open the CSV file in read mode ('r')
data = []
with open('data.csv', 'r') as file:
    # Create a csv.reader object
    reader = csv.reader(file)

    # Optionally, read the header row
    header = next(reader)
    # print(f"Header: {header}")

    # Iterate over the remaining rows
    for row in reader:
        data.append([float(r) for r in row])
data = np.array(data)
```

```python
import matplotlib.pyplot as plt
task = data[:,0]
makespan = data[:,1]
cost = data[:,2]
replan = data[:,3]
service_time = data[:,4]

fig,ax = plt.subplots(2,2,figsize=(10,6))
ax[0,0].plot(task,makespan,lw=3)
ax[0,1].plot(task,cost,lw=3)
ax[1,0].plot(task,replan,lw=3)
ax[1,1].plot(task,service_time,lw=3)

ax[0,0].set_ylabel('Makespan',fontsize=16)
ax[0,1].set_ylabel('Cost',fontsize=16)
ax[1,0].set_ylabel('Number of Replan',fontsize=16)
ax[1,1].set_ylabel('Service Time',fontsize=16)
ax[1,0].set_xlabel('Number of Tasks',fontsize=16)
ax[1,1].set_xlabel('Number of Tasks',fontsize=16)
fig.suptitle("Varying Task Number",fontsize=20)
fig.tight_layout()
```
